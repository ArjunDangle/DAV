```python
import seaborn as sns
import matplotlib.pyplot as plt

# Load the Iris dataset
df = sns.load_dataset('iris')

# 1. Relationship Plots
# Scatter Plot
sns.scatterplot(x='sepal_length', y='sepal_width', hue='species', data=df)
plt.title('Scatter Plot: Sepal Length vs Sepal Width')
plt.show()

# Line Plot
sns.lineplot(x='sepal_length', y='sepal_width', hue='species', data=df)
plt.title('Line Plot: Sepal Length vs Sepal Width')
plt.show()

# 2. Categorical Plots
# Count Plot
sns.countplot(x='species', data=df)
plt.title('Count of Each Species')
plt.show()

# Violin Plot
sns.violinplot(x='species', y='petal_length', data=df)
plt.title('Violin Plot: Petal Length by Species')
plt.show()

# 3. Distribution Plots
# Histogram
sns.histplot(df['petal_length'], kde=False)
plt.title('Histogram of Petal Length')
plt.show()

# KDE Plot
sns.kdeplot(data=df, x='petal_length', hue='species', fill=True)
plt.title('KDE Plot of Petal Length by Species')
plt.show()

# 4. Matrix Plots
# Heatmap of correlations
corr = df.drop(columns='species').corr()
sns.heatmap(corr, annot=True, cmap='coolwarm')
plt.title('Heatmap of Feature Correlations')
plt.show()

# Cluster Map
sns.clustermap(corr, annot=True, cmap='viridis')
plt.title('Cluster Map of Feature Correlations')
plt.show()

# 5. Multiple Variable Plots
# Pair Plot
sns.pairplot(df, hue='species')
plt.suptitle('Pair Plot of Iris Dataset', y=1.02)
plt.show()

# Joint Plot
sns.jointplot(x='sepal_length', y='sepal_width', data=df, kind='scatter', hue='species')
plt.suptitle('Joint Plot: Sepal Length vs Sepal Width', y=1.02)
plt.show()

# 6. Regression Plots
# Residual Plot
sns.residplot(x='sepal_length', y='sepal_width', data=df)
plt.title('Residual Plot: Sepal Length vs Sepal Width')
plt.show()

# Regression Plot
sns.regplot(x='sepal_length', y='sepal_width', data=df)
plt.title('Regression Plot: Sepal Length vs Sepal Width')
plt.show()
```
