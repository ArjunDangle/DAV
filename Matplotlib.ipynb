```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.datasets import load_iris
from mpl_toolkits.mplot3d import Axes3D

# Load Iris dataset
iris = load_iris()
X = iris.data
y = iris.target
feature_names = iris.feature_names
target_names = iris.target_names

# 1. Basic Plots - Line, Scatter, and Pie
def basic_plots():
    # Line Plot (Petal Length over samples)
    plt.plot(X[:, 2], label='Petal Length')
    plt.title('Line Plot of Petal Length')
    plt.xlabel('Sample Index')
    plt.ylabel('Petal Length (cm)')
    plt.legend()
    plt.show()

    # Scatter Plot (Sepal Length vs Sepal Width)
    plt.scatter(X[:, 0], X[:, 1], c=y, cmap='viridis')
    plt.title('Scatter Plot: Sepal Length vs Sepal Width')
    plt.xlabel('Sepal Length (cm)')
    plt.ylabel('Sepal Width (cm)')
    plt.colorbar(label='Species')
    plt.show()

    # Pie Chart (Distribution of classes)
    counts = np.bincount(y)
    plt.pie(counts, labels=target_names, autopct='%1.1f%%', startangle=90)
    plt.title('Species Distribution')
    plt.show()

# 2. Statistical and Distribution Plots
def static_distribution_plots():
    # Histogram (Petal Length)
    plt.hist(X[:, 2], bins=20, color='skyblue', edgecolor='black')
    plt.title('Histogram of Petal Length')
    plt.xlabel('Petal Length (cm)')
    plt.ylabel('Frequency')
    plt.show()

    # Box Plot (All features)
    plt.boxplot(X, labels=feature_names, vert=False)
    plt.title('Box Plot of All Features')
    plt.show()

    # Violin Plot (requires transformation)
    plt.violinplot([X[y == i, 2] for i in range(3)], showmeans=True)
    plt.xticks([1, 2, 3], target_names)
    plt.title('Violin Plot of Petal Length by Species')
    plt.show()

# 3. Categorical Plots (Bar & Grouped)
def categorical_plots():
    counts = np.bincount(y)
    # Bar Plot
    plt.bar(target_names, counts, color='green')
    plt.title('Bar Plot of Class Counts')
    plt.ylabel('Count')
    plt.show()

    # Grouped Bar Chart (Average feature values per class)
    avg = np.array([X[y == i].mean(axis=0) for i in range(3)])
    x = np.arange(len(feature_names))
    width = 0.2
    for i in range(3):
        plt.bar(x + width * i, avg[i], width=width, label=target_names[i])
    plt.xticks(x + width, feature_names, rotation=45)
    plt.title('Grouped Bar Chart of Mean Feature Values')
    plt.ylabel('Value')
    plt.legend()
    plt.tight_layout()
    plt.show()

# 4. Multi-series, Subplots, Twin Axes
def multiseries_subplot_plots():
    x = np.arange(len(X))

    # Multi-line Plot
    plt.plot(X[:, 0], label='Sepal Length')
    plt.plot(X[:, 1], label='Sepal Width')
    plt.title('Multiline Plot')
    plt.xlabel('Sample Index')
    plt.ylabel('cm')
    plt.legend()
    plt.show()

    # Subplots
    plt.figure(figsize=(8, 6))
    plt.subplot(2, 1, 1)
    plt.plot(X[:, 2], color='r')
    plt.title('Petal Length')
    plt.subplot(2, 1, 2)
    plt.plot(X[:, 3], color='b')
    plt.title('Petal Width')
    plt.tight_layout()
    plt.show()

    # Twin Axes
    fig, ax1 = plt.subplots()
    ax2 = ax1.twinx()
    ax1.plot(X[:, 0], 'g-', label='Sepal Length')
    ax2.plot(X[:, 2], 'b-', label='Petal Length')
    ax1.set_ylabel('Sepal Length (cm)', color='g')
    ax2.set_ylabel('Petal Length (cm)', color='b')
    plt.title('Twin Axes Plot')
    plt.show()

# 5. 3D and Advanced Plots
def advanced_3d_plots():
    # 3D Scatter Plot
    fig = plt.figure()
    ax = fig.add_subplot(111, projection='3d')
    ax.scatter(X[:, 0], X[:, 1], X[:, 2], c=y, cmap='viridis')
    ax.set_xlabel('Sepal Length')
    ax.set_ylabel('Sepal Width')
    ax.set_zlabel('Petal Length')
    plt.title('3D Scatter Plot')
    plt.show()

# 6. Special Plots
def special_plots():
    # Heatmap (correlation)
    corr = np.corrcoef(X.T)
    plt.imshow(corr, cmap='coolwarm', interpolation='nearest')
    plt.colorbar()
    plt.xticks(range(4), feature_names, rotation=45)
    plt.yticks(range(4), feature_names)
    plt.title('Feature Correlation Heatmap')
    plt.tight_layout()
    plt.show()

    # Polar Plot (synthetic for demonstration)
    theta = np.linspace(0, 2 * np.pi, 150)
    r = np.abs(np.sin(theta))
    plt.subplot(projection='polar')
    plt.plot(theta, r)
    plt.title('Polar Plot')
    plt.show()

    # Error Bars
    means = X.mean(axis=0)
    stds = X.std(axis=0)
    x = np.arange(len(feature_names))
    plt.errorbar(x, means, yerr=stds, fmt='o', color='purple')
    plt.xticks(x, feature_names, rotation=45)
    plt.title('Feature Means with Error Bars')
    plt.ylabel('cm')
    plt.tight_layout()
    plt.show()

# Run all
basic_plots()
static_distribution_plots()
categorical_plots()
multiseries_subplot_plots()
advanced_3d_plots()
special_plots()
```
